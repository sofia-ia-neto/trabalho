```python
"vamos agora trabalhar sobre as features do ficheiro genbank"
"para obter as zonas codificantes da proteina"
```

```python
from Bio import SeqFeature
from Bio import SeqIO

record = SeqIO.read("FANCF.gb", "genbank")

featcds=[]
for i in range(len(record.features)):
    if record.features[i].type == "CDS":
        featcds.append(i)

for k in featcds:
    print(record.features[k].location)
for k in featcds:
    print("\n")
    print ("zona codificante:", record.features[k].extract(record.seq))
    
```
